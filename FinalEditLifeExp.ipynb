### Create models with different degrees to compare the R-squared and RMSE (using all features)

```python
def modelcreation(number):
 #transform the polynomial features
    poly = PolynomialFeatures(degree = number)

    y = df['Life expectancy ']
    x = df.drop('Life expectancy ', axis=1)
    
    features_poly = poly.fit_transform(x)
    
    print('Polynomial with degree: ',number)

    #split to train and test data
    x_train, x_test, y_train, y_test = train_test_split(features_poly, y, train_size=0.9,test_size=0.1, random_state=42)
    
    #fit in a linear regression
    poly_model = LinearRegression()
    poly_model.fit(x_train, y_train)
    
    #prediction
    poly_predictions = poly_model.predict(x_test)
    
    #Evaluation
    # R-squared
    r2_train = poly_model.score(x_train, y_train)
    r2_test = poly_model.score(x_test, y_test)

    # RMSE
    rmse = mean_squared_error(y_test, poly_predictions)
    
    print("R^2 in training set is:", r2_train)
    print("R^2 in test set is:", r2_test)
    print('RMSE is :', rmse**0.5)
    print('\n')

#Looping degree from 1 to 10 to find the one produce the best R-squared
for deg in range(1,6):
    modelcreation(deg)
```

## Support Vector Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler



# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)

# Standardize the features using StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

# Create an SVR model
svr = SVR(kernel='rbf')  # You can choose different kernels like 'linear', 'poly', 'rbf', etc.

# Train the SVR model on the training data
svr.fit(x_train, y_train)

# Make predictions on the test data
y_pred = svr.predict(x_test)

# Evaluate the model's performance (e.g., using mean squared error or other metrics)
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```

## Using Logistic Regression as a classification Algorithm

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


df['LifeExpectancyCategory'] = pd.cut(df['Life expectancy '], bins=[0, 60, 75, df['Life expectancy '].max()], labels=['Low', 'Medium', 'High'])

# Select the target variable and features
y = df['LifeExpectancyCategory']
x = df.drop(['Life expectancy ', 'LifeExpectancyCategory'], axis=1)

# Split the data into training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.9, test_size=0.1, random_state=42)

# Create a Logistic Regression model
logistic_regression = LogisticRegression()

# Train the model on the training data
logistic_regression.fit(x_train, y_train)

# Make predictions on the test data
y_pred = logistic_regression.predict(x_test)

# Evaluate the classification model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Accuracy:", accuracy)
print(report)
```

## Kmeans Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


features = df[['Adult Mortality','Life expectancy ']]

# Standardize the features (scaling)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)


K = 4

# Create a K-Means model
kmeans = KMeans(n_clusters=K, random_state=42)

# Fit the K-Means model to the scaled features
kmeans.fit(scaled_features)

# Get cluster assignments for each data point
cluster_assignments = kmeans.labels_

# Add the cluster assignments to the DataFrame
df['Cluster'] = cluster_assignments

# Visualize the results
plt.scatter(df['Life expectancy '], df['Adult Mortality'], c=cluster_assignments, cmap='copper_r')
plt.xlabel('Life expectancy ')
plt.ylabel('Adult Mortality')
plt.title(f'K-Means Clustering (K={K})')
plt.show()

# You can now explore and analyze the results based on the clusters
```

## Ridge and Lasso After Scaling

```python
X_train, X_test, y_train, Y_test = train_test_split(feature_ss, y, test_size=0.1, random_state = 42)

from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV, KFold
import numpy as np

# Assuming X_train and y_train are your training data
param_grid = {'alpha': np.logspace(-4,4,9)}  # Define a range of alpha values to search

# Create Ridge and Lasso regression models
ridge_model = Ridge()
lasso_model = Lasso()

# Create a k-fold cross-validation object
kf = KFold(n_splits=5, shuffle=True, random_state=42)  # You can adjust the number of folds (n_splits) as needed

# Perform grid search for Ridge
ridge_grid = GridSearchCV(estimator=ridge_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=kf)
ridge_grid.fit(X_train, y_train)

# Perform grid search for Lasso
lasso_grid = GridSearchCV(estimator=lasso_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=kf)
lasso_grid.fit(X_train, y_train)

# Get the best alpha values for Ridge and Lasso
best_alpha_ridge = ridge_grid.best_params_['alpha']
best_alpha_lasso = lasso_grid.best_params_['alpha']

# Train Ridge and Lasso models with the best alpha values
best_ridge_model = Ridge(alpha=best_alpha_ridge)
best_lasso_model = Lasso(alpha=best_alpha_lasso)

# Fit the models to the data
best_ridge_model.fit(X_train, y_train)
best_lasso_model.fit(X_train, y_train)

# Now you can use the best Ridge and Lasso models for prediction
print("Best Ridge Alpha:", best_alpha_ridge)
print("Ridge MSE:", -ridge_grid.best_score_)  # Use -ridge_grid.best_score_ to get the positive MSE value

print("Best Lasso Alpha:", best_alpha_lasso)
print("Lasso MSE:", -lasso_grid.best_score_)  # Use -lasso_grid.best_score_ to get the positive MSE value
```

```python
# Let's create a binary classification task for simplicity
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
df['Life Expectancy Category'] = np.where(df['Life expectancy '] > df['Life expectancy '].median(), 1, 0)

# Select features and target variable
X = df[['GDP', 'Schooling', 'Alcohol']]
y = df['Life Expectancy Category']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Train a simple classifier (e.g., Decision Tree)
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = classifier.predict(X_test)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Visualize the confusion matrix using seaborn heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Print accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
